```python
import pandas as pd

def min_max_medie(dataframe, coloana):
    values=[]
    values.append(min(dataframe[coloana].dropna()))
    values.append(max(dataframe[coloana].dropna()))
    values.append(dataframe[coloana].mode()[0])
    return values

df_employees = pd.read_csv('employees.csv',delimiter=',',header=0)

for coloana in df_employees.columns:
    values = min_max_medie(df_employees,coloana)
    print(coloana,':')
    print("min: ",values[0])
    print("max: ",values[1])
    print("medie: ",values[2])
    print()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


df_employees = pd.read_csv('employees.csv',delimiter=',',header=0)

plt.figure()
df_employees['Salary'].hist(bins=20, color='skyblue', edgecolor='black')
plt.title('Distribuția salariilor')
plt.xlabel('Salariu')
plt.ylabel('Număr de angajați')
plt.grid(False)
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt


df_employees = pd.read_csv('employees.csv',delimiter=',',header=0)

plt.figure(figsize=(10, 6))
df_employees.boxplot(column='Salary', by='Team', figsize=(10,6), grid=False)
plt.title('Distribuția salariilor pe echipă')
plt.xlabel('Echipă')
plt.ylabel('Salariu')
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd

df_employees = pd.read_csv('employees.csv')

q_low = df_employees['Salary'].quantile(0.01)
q_high = df_employees['Salary'].quantile(0.99)

outliers_low = df_employees[df_employees['Salary'] < q_low]
outliers_high = df_employees[df_employees['Salary'] > q_high]

print("Angajații cu cele mai mici salarii:")
print(outliers_low)

print("\nAngajații cu cele mai mari salarii:")
print(outliers_high)
```
